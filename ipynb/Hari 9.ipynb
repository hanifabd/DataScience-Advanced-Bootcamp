```python
topic = 'jouska'
since_date = '2020-07-25'
```

```python
ns = topic + " -filter:retweets"
```

```python
tweets = tp.Cursor(api.search, q=ns, lang="id", since=since_date, tweet_mode="extended").items(50)
```

```python
items = []

for tweet in tweets:
    item = []
    item.append (tweet.user.screen_name)
    item.append (tweet.user.location)
    item.append (' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet.full_text).split()))
    items.append(item)
```
