```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
alamat = 'https://en.wikipedia.org/wiki/List_of_brightest_stars'
html = urlopen(alamat)
data = bs(html)
```

```python
table = data.findAll('table', {'class':'wikitable'})[0]
```

```python
rows = table.findAll('tr')
```

```python
df = []
for row in rows:
    aRow = []
    for cell in row.findAll(['th','td']):
        data = cell.get_text()
        aRow.append(data.strip())
    df.append(aRow)
```

```python
df = pd.DataFrame(df)
```

```python
head = df.iloc[0]
df.columns = head
df = df[1:]
```

```python
df.reset_index(drop=True)
```

```python
df['Spectral class'].iloc[0] = df['Distance (ly)'].iloc[0]
df['Distance (ly)'].iloc[0] = df['part 2'].iloc[0]
df['part 2'].iloc[0] = ''
```
