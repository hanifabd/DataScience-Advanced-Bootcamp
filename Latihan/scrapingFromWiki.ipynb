```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
alamat = 'https://en.wikipedia.org/wiki/Comparison_of_text_editors'
html = urlopen(alamat)
data = BeautifulSoup(html, 'html.parser')
```

```python
table = data.findAll('table', {'class':'wikitable'})[1]
```

```python
rows = table.findAll('tr')
```

```python
csvRows = []
for row in rows:
    csvRow = []
    for cell in row.findAll(['td','th']):
        csvRow.append(cell.get_text())
    csvRows.append(csvRow)
```

```python
csvRows = pd.DataFrame(csvRows)
```

```python
csvRows
```

```python
new_header = csvRows.iloc[0]
new_header
```

```python
csvRows = csvRows[0]
```

```python
csvAlts = []
for row in rows:
    csvAlt = []
    for img in row.findAll('img', alt=True):
        csvAlt.append(img['alt'])
    csvAlts.append(csvAlt)
```

```python
csvAlts
```

```python
csvAlts = pd.DataFrame(csvAlts)
```

```python
csvAlts
```

```python
csvRows = pd.DataFrame(csvRows)
```

```python
csvRows
```

```python
result = pd.concat([csvRows, csvAlts], axis=1, sort=False)
```

```python
result
```

```python
df = result[1:]
df.columns = ['Text Editor', 'Microsoft Windows', 'macOS', 'Linux', 'BSD', 'Unix', 'OpenVMS']
```
